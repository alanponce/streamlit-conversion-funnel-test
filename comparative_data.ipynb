## Using Conversion_rate_user_info

```python
prueba_st = df_streamlit [ df_streamlit["EventDateTime_soft"].dt.month == 11]
prueba_st
```

```python
len( prueba_st[ prueba_st["FundingIn2weeks_accountIF"] ==1  ].index)
```

```python
# registros positvos de FundingIn2weeks_accountIF / total del df en noviembre
len( prueba_st[ prueba_st["FundingIn2weeks_accountIF"] ==1  ].index) / len(prueba_st.index)
```

```python
result = df_streamlit.groupby([df_streamlit['EventDateTime_soft'].dt.to_period("M")]).agg(
    total_rows=pd.NamedAgg(column='EventDateTime_soft', aggfunc='size'),
    count_values_positives=pd.NamedAgg(column='FundingIn2weeks_accountIF', aggfunc=lambda x: sum(x == 1))
).reset_index()

result["ConversionRate"] = ( result["count_values_positives"] / result["total_rows"] ) * 100

result
```

```python
result["EventType"] = [df_streamlit.dropna()["EventType_accountIF"].unique()[0]] * len(result.index)
result
```

## Using Conversion_rate_user_info, without na in platform, age and income

```python
prueba_st_without_na = prueba_st.dropna(subset=["Platform", "User.M.Age.N", "User.M.Income.N"], how='any')
```

```python
prueba_st_without_na
```

```python
len( prueba_st_without_na[ prueba_st_without_na["FundingIn2weeks_accountIF"] ==1  ].index) / len(prueba_st_without_na.index)
```

```python
df_streamlit_with_na = df_streamlit.dropna(subset=["Platform", "User.M.Age.N", "User.M.Income.N"], how='any')
```

```python
result = df_streamlit_with_na.groupby([df_streamlit_with_na['EventDateTime_soft'].dt.to_period("M")]).agg(
    total_rows=pd.NamedAgg(column='EventDateTime_soft', aggfunc='size'),
    count_values_positives=pd.NamedAgg(column='FundingIn2weeks_accountIF', aggfunc=lambda x: sum(x == 1))
).reset_index()

result["ConversionRate"] = ( result["count_values_positives"] / result["total_rows"] ) * 100

result
```

```python
result["EventType"] = [prueba_st_without_na.dropna()["EventType_accountIF"].unique()[0]] * len(result.index)
result
```

```python
df_filter =  df_streamlit[df_streamlit['User.M.Age.N'].between( 17, 69)]
df_filter =  df_filter[df_filter['User.M.Income.N'].between( 12500.00, 184000.00)]
df_filter =  df_filter[df_filter['Platform'] == "iOS"]
```

```python
result = df_filter.groupby([df_filter['EventDateTime_soft'].dt.to_period("M")]).agg(
    total_rows=pd.NamedAgg(column='EventDateTime_soft', aggfunc='size'),
    count_values_positives=pd.NamedAgg(column='FundingIn2weeks_accountIF', aggfunc=lambda x: sum(x == 1))
).reset_index()

result["ConversionRate"] = ( result["count_values_positives"] / result["total_rows"] ) * 100

result
```
