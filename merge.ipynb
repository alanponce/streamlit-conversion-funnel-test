```python
import pandas as pd
import numpy as np

```

```python
df_quicksight = pd.read_csv("data/concatenated_df_2.csv")
df_streamlit = pd.read_csv("data/users_daily_2023_12_18.csv")


#df_quicksight[ 'EventDateTime_soft' ] = pd.to_datetime(df_quicksight[ 'EventDateTime_soft'], format="%Y-%m-%dT%H:%M:%S%z")

df_quicksight['EventDateTime_soft'] = pd.to_datetime(df_quicksight['EventDateTime_soft'], format='ISO8601')

```

```python
csv_merge = pd.merge(df_quicksight, df_streamlit, how="left", left_on= "UserId", right_on="userId")
csv_merge
```

```python
csv_merge['dateOfBirth'] = pd.to_datetime(csv_merge['dateOfBirth'], utc=True)
current_date = pd.to_datetime('today', utc=True)
csv_merge['age'] = np.floor((current_date - csv_merge['dateOfBirth']) / np.timedelta64(1, 'Y'))
```

```python
csv_merge.to_csv("data/merge_data_v2.csv", index=False)
```
